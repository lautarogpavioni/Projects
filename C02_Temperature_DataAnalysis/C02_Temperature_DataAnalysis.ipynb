```python
world_1 = pd.DataFrame(df.groupby('Year')['5-yr smooth'].mean())
world_2 = pd.DataFrame(df.groupby('Year')['Annual Mean'].mean())
world = pd.concat([world_1, world_2], axis=1)
world=world.rename(columns={'5-yr smooth': '5-yr smooth T', 'Annual Mean': 'Annual Mean T'})
world
```

```python
plt.figure(figsize=(8,5))
# arg['5-yr smooth'].plot()
sns.scatterplot(data = world, x = 'Year', y = '5-yr smooth T')
plt.show()
```

```python
# emission by country
ebc = pd.read_csv('GCB2022v27_MtCO2_flat.csv')
ebc
```

```python
ebc[ebc['Total'] != 0]
```

global emission from years > 1900

```python
ebc = ebc.query('Year >= 1901')
g_e = ebc[ebc['Country'] == 'Global']
g_e
```

We want both dataframes to have same dimension:

```python
world = world[:-1]
print(g_e.shape)
world.shape
```

Concatenate both data frames:

```python
g_e.index = g_e['Year']
df = pd.concat( [g_e, world], axis = 1)
df = df.drop(['Year', 'Country', 'ISO 3166-1 alpha-3', 'Other'], axis=1)
```

```python
df.index
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df['5-yr smooth T'] = scaler.fit_transform(df['5-yr smooth T'].values.reshape(-1,1))
df['Annual Mean T'] = scaler.fit_transform(df['Annual Mean T'].values.reshape(-1,1))
df['Total'] =  scaler.fit_transform(df['Total'].values.reshape(-1,1))
```

```python
plt.figure(figsize=(8,5))
# arg['5-yr smooth'].plot()
sns.scatterplot(data = df, x = 'Year', y = 'Total', label = "Emisiones")
sns.scatterplot(data = df, x = 'Year', y = 'Annual Mean T', label = "mean Temperature", alpha = 0.5)
sns.scatterplot(data = df, x = 'Year', y = '5-yr smooth T', label = "5-year smooth Temperature")
plt.show()
```
